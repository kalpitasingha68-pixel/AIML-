Data Visualization

```python
# Matplotlib Bar Chart

import math
import datetime
import matplotlib.pyplot as plt

students =["Rahul", "Madhu", "Sara", "Amit"]
marks = [85, 92, 78, 88]

plt.figure(figsize=(20, 10))
plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
# Line Chart

import matplotlib.pyplot as plt

days =[1, 2, 3, 4, 5]
sales =[200, 250, 300, 280, 350]

plt.figure(figsize=(8, 5))
plt.plot(days, sales, marker='o', color='green')
plt.title("Daily Sales")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

```python
# Histogram

import matplotlib.pyplot as plt

marks =[85, 92, 78, 88, 90, 76, 84, 91, 87, 89]

plt.figure(figsize=(8, 5))
plt.hist(marks, bins=16, edgecolor='black')
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
# Seaboen Count Plot

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data= pd.DataFrame({
    "Department": ["AI", "AI", "ML", "DS", "DS", "AI", "ML"]
})
plt.figure(figsize=(6, 8))
sns.countplot(x="Department", data=data, color="green")
plt.title("Department Count")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
# Seaborn Heatmap

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "Marks": [85, 92, 78, 88],
    "Attendence": [90, 95, 80, 85],
    "StudyHours": [3, 5, 2, 4]
})

plt.figure(figsize=(6, 4))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm",)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Filter Data

import pandas as pd

data ={
    "Name": ["Alice", "Madhu", "Sara", "Amit"],
    "Marks": [85, 92, 78, 88]
}
df = pd.DataFrame(data)
high_marks = df[df["Marks"] > 80]
print(high_marks)
```

```python
# Missing Values example

import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Madhu", "Sara", "Amit"],
    "Marks": [85, np.nan, 78, 88],
    "Attendance": [90, 95, np.nan, 85]
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)

print("\nMissing Values:")
print(df.isnull())

df["Marks"].fillna(df["Marks"].mean(), inplace=True)
df["Attendance"].fillna(df["Attendance"].mean(), inplace=True)

print("\nCleaned Data:")
print(df)
```

```python
# Remove Duplicates

import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Madhu", "Sara", "Amit"],
    "Marks": [85, 76, 78, 88],
    "Attendance": [90, 95, 80, 85]
}
df = pd.DataFrame(data)
print("Before removing duplicates:")
print(df)

df = df.drop_duplicates()

print("\nAfter removing duplicates:")
print(df)
```
