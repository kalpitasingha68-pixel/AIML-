String and its method

```python
text = "hello world"
print(text.upper())
print(text.lower())
print(text.title())
print(text.capitalize())
print(text.replace("world", "Python"))
print(text.find("world"))
print(len(text))
```

List and Its Method

```python
numbers =[10, 20, 30]

numbers.append(40) #add at end
print(numbers)

numbers.insert(1, 15) #add at index 1
print(numbers)

numbers.remove(20) #remove value
print(numbers)

numbers.pop() #remove last item
print(numbers)

numbers.sort() #sort list
print(numbers)

numbers.reverse() #reverse list
print(numbers)
```

Another list simple example

```python
fruits = ["apple", "banana", "mango"]

fruits.append("orange")
print(fruits)

fruits.remove("banana")
print(fruits)

print(len(fruits))
```

Nested Loop


```python
for i in range (1, 4):
    for j in range(1, 4):
     print(i, j)
```

Break


```python
for i in range (1, 10):
  if i ==5:
     break
  print(i)
```

Continue

```python
for i in range (1, 6):
  if i ==5:
     continue
  print(i)
```

User defined Functions


```python
def greet():
  print("Welcome to Python class")

greet()
```

```python
def greet(name):
  print(f"Welcome {name}")

greet("Alex")
greet("Miota")
greet("kalpi")
```

```python
# sum of two number using function
def add(a, b):
  return a + b

result = add(10, 20)
print(f"The sum is: {result}")
```

```python
def square(num):
  return num * num

x = 5 # Assign a numerical value to x
print("Square:", square(x))
```
