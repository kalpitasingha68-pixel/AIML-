```python
import numpy as np

def convert(x):
    if isinstance(x, str):
        x = x.strip()
        if "-" in x:
            try:
                a, b = x.split("-") # Corrected typo: used parentheses for .split()
                return (float(a) + float(b)) / 2
            except ValueError: # Corrected indentation, added specific exception
                return np.nan
        else: # Corrected indentation for else branch
            try:
                return float(x)
            except ValueError:
                return np.nan
    else: # Corrected indentation for else branch
        try:
            return float(x)
        except ValueError:
            return np.nan
    return np.nan # Corrected indentation
```

```python
df['total_sqft']=df['total_sqft'].apply(convert)
df.dropna(inplace=True)
```
